***Veri bilgilendirme dosyasini parse etme***

```data_description.txt``` dosyasini sozluk olarak parse edip ```data_description.json``` dosyasina kaydeder.


````python
import json
dict = {}

# txt dosyayi alip sozluk olarak parse edelim
with open('data_description.txt') as f:
    lines = f.readlines() # list containing lines of file
    features = [] # To store feature names

    for line in lines:
        line = line.strip() # remove leading/trailing white spaces
        if line:
            if ':' in line:
                features = [item.strip() for item in line.split(':')]
                d = {'description': features[1], 'values': {}}
                dict[features[0]]=d
                featurename = features[0]
            else:
                values = [item for item in line.split('\t')]                        
                dict[featurename]['values'][values[0]]= values[1]

# parse ettigimiz sozlugu json dosyasi olarak kaydedelim
with open('data_description.json', 'w') as fp:
    json.dump(dict, fp)

# son olarak test etmek icin json dosyasini okuyup yine bir sozluge yukleyelim
with open('data_description.json', 'r') as f:
    data_dict = json.load(f)

data_dict
````
