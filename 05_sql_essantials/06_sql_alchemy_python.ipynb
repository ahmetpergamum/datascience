# Ubuntu 18.04 icin onemli not

- SQLAlchemy modulu db baglantisi yaparken ```caching_sha2_password``` hatasi veriyor cunku MySQL 8 ve uzerinde bu policy varsayilan olarak geliyor. Bizim kullandigimiz programda ise kullanici ```mysql_native_password``` policy ile baglanmaya calisiyor.
- Bunu cozmek icin Ubuntu 18.04 de;
    - ```/etc/my.conf``` dosyasina asagidaki satirlari eklemeliyiz, ustteki ```[mysqld]``` satirini unutmayin yoksa mysql baslarken hata veriyor. Konuyla ilgili [link](https://dev.mysql.com/doc/refman/8.0/en/upgrading-from-previous-series.html#upgrade-caching-sha2-password).
    
        ```[mysqld]```
      
        ```default-authentication-plugin=mysql_native_password```
    - pip3 ile asagidaki modulleri kurmaliyiz. (asagidaki pip komutlari pip3 un aliasi seklinde calismaktadir.)
    
        ```pip install sqlalchemy```
        
        ```pip install mysql-connector-python```
    - ```mysql-connector-python``` yerine ```mysql-connector``` kurdugumuzda konfigurasyon dosyasi degisikligi yapsak bile ayni hatayi almaya devam ediyoruz. Konuyla ilgili [stackoverflow](https://stackoverflow.com/a/53561512)'a goz atilabilir.

## Q3
```sakila``` şemasında yer alan ```customer``` tablosundan ```first_name``` ve ```last_name``` alanlarını sorgulayarak ekrana döktüren python kodu yazınız.

---

````python
from sqlalchemy import create_engine

# engine tanimlamasinda kullanici, parola ve db bilgileri girilir
engine = create_engine('mysql+mysqlconnector://odev:Odev1234!@localhost/sakila')

# connection nesnesi olusturulur
connection = engine.connect()
# connection nesnesinin execute metodu ile istenen sql sorgusu calistirilir
result = connection.execute("select first_name, last_name from customer")
for row in result:
    #print(row)
    print('Adi Soyadi: {} {}'.format(row['first_name'], row['last_name']))
    
# bağlantıyı kapatmayı unutmamamız lazım
connection.close()
````
