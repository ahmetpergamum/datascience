# Temel İstatistik Kavramları

### Q1
numpy kütüphanesini kullanarak üreteceğiniz 3 farklı veri aracılığı ile merkezi eğilim ve dağılım ölçümlerini hem yukarıda verilen matematiksel formülle hem de python'da hazır bulunan (built-in) kodlar aracılığıyla analiz edin!!!

```python
import numpy as np
import statistics
import matplotlib.pyplot as plt
import math
import warnings
warnings.filterwarnings('ignore')   

# mode calculation without error for multiple modes
def calcMode(arr):
    # unique values and counts of them, np.unique returns two array
    (values, counts) = np.unique(arr, return_counts=True)
    # get the index of most frquent value 
    ind = np.argmax(counts)
    # retrieve the most frequent element
    return values[ind]

def calcVariance(arr):
    mean = sum(arr)/len(arr)
    return sum([(i - mean)**2 for i in arr]) / (len(arr)-1)

def calcStdDev(arr):
    return math.sqrt(calcVariance(arr))
    
def calcStdErr(arr):
    return calcStdDev(arr) / math.sqrt(len(arr))
    
# create normal, exponential and grade arrays
normal = np.random.normal(10, 10, 1000)
expo = np.exp(np.arange(0,100,0.1))
grades = np.array([23,35,56,65,89,76,65,99,45,56,73,74,45,54,44,77,87,89,87,96,99,100,0,53])

print('\nCalculation with python libraries\n----------------------------------\n')
# Mean of arrays:
print('Mean of arrays \n----------------------------------')
print('Normal distribution \t: {:.2f}'.format(np.mean(normal)))
print('Exponential distribution \t: {:.2f}'.format(np.mean(expo)))
print('Grades \t\t\t: {:.2f}'.format(np.mean(grades)))

print('----------------------------------\nMedian of arrays \n----------------------------------')
print('Normal distribution \t: {:.2f}'.format(np.median(normal)))
print('Exponential distribution \t: {:.2f}'.format(np.median(expo)))
print('Grades  \t\t\t: {:.2f}'.format(np.median(grades)))

print('----------------------------------\nMode of arrays \n----------------------------------')
print('Normal distribution \t: ', end=' ')
try:
    print(statistics.mode(normal))
except statistics.StatisticsError:
    print('Variables are uniqe and cannot calculate mode!')
print('Exponential distribution \t: ', end=' ')
try:
    print(statistics.mode(expo))
except statistics.StatisticsError:
    print('Variables are uniqe and cannot calculate mode!')
    
print('Grades \t\t\t: ',calcMode(grades))

print('----------------------------------\nVariance of arrays \n----------------------------------')
print('Normal distribution \t: {:.2f}'.format(np.var(normal, ddof=1)))
print('Exponential distribution \t: \n{:.2f}'.format(np.var(expo, ddof=1)))
print('Grades \t\t\t: {:.2f}'.format(np.var(grades, ddof=1)))

print('----------------------------------\nStandard deviation of arrays \n----------------------------------')
print('Normal distribution \t: {:.2f}'.format(np.std(normal, ddof=1)))
print('Exponential distribution \t: {:.2f}'.format(np.std(expo, ddof=1)))
print('Grades \t\t\t: {:.2f}'.format(np.std(grades, ddof=1)))

print('----------------------------------\nStandard error of arrays \n----------------------------------')
print('Normal distribution \t: {:.2f}'.format(np.std(normal, ddof=1) / np.sqrt(len(normal))))
print('Exponential distribution \t: {:.2f}'.format(np.std(expo, ddof=1)/ np.sqrt(len(expo))))
print('Grades \t\t\t: {:.2f}'.format(np.std(grades, ddof=1) / np.sqrt(len(grades))))

print('\n\nCalculation manually\n----------------------------------\n')
print('Variance of arrays \n----------------------------------')
print('Normal distribution \t: {:.2f}'.format(calcVariance(normal)))
print('Exponential distribution \t: \n{:.2f}'.format(calcVariance(expo)))
print('Grades \t\t\t: {:.2f}'.format(calcVariance(grades)))
print('----------------------------------\nStandard deviation of arrays \n----------------------------------')
print('Normal distribution \t: {:.2f}'.format(calcStdDev(normal)))
print('Exponential distribution \t: {:.2f}'.format(calcStdDev(expo)))
print('Grades \t\t\t: {:.2f}'.format(calcStdDev(grades)))

print('----------------------------------\nStandard error of arrays \n----------------------------------')
print('Normal distribution \t: {:.2f}'.format(calcStdErr(normal)))
print('Exponential distribution \t: {:.2f}'.format(calcStdErr(expo)))
print('Grades \t\t\t: {:.2f}'.format(calcStdErr(grades)))

print('Graphs of arrays')
plt.figure(figsize=(36,8))
plt.subplot(131)
plt.hist(normal, bins=100, color='lightblue')
plt.title('Normal Dagilim')
plt.subplot(132)
plt.hist(expo, bins=100, color='purple')
plt.title('Exponensiyel Dagilim')
plt.subplot(133)
plt.hist(grades, bins=100, color='lightgreen')
plt.title('Notlarin Dagilimi')

plt.show()
```
