### Bar chart of GDP and Population by Regions.

Data is retrieved from https://www.kaggle.com/fernandol/countries-of-the-world

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# make dataframe from csv file
df = pd.read_csv('countries_of_the_world.csv')
r = df.groupby('Region').mean().index
# strip the region names
regions = []
for region in r:
    regions.append(region.strip())
# make gdps an array
gdps = np.array(df.groupby('Region').mean().loc[:,'GDP ($ per capita)'])
# make sum of population an array
population = np.array(df.groupby('Region').sum().loc[:,'Population'])
# make a new dataframe to sort by gdp and population
dt = pd.DataFrame()
dt['Region'] = regions
dt['GDP'] = gdps
dt['Population'] = population

# sort by gdp
sorted_gdp_region = dt.sort_values(by='GDP').loc[:,'Region']
sorted_gdp = dt.sort_values(by='GDP').loc[:,'GDP']
# set the figure size
plt.figure(figsize=(22, 8))
# left plot settings
plt.subplot(121)
plt.bar(sorted_gdp_region,sorted_gdp ,color='lightblue')
# rotate the labels in x axis
plt.xticks(rotation=65)
plt.title("GDP by Region")
plt.ylabel("GDP ($ per capita)")

# sort by population
sorted_pop_region = dt.sort_values(by='Population').loc[:,'Region']
sorted_pop = dt.sort_values(by='Population').loc[:,'Population']
# right plot settings
plt.subplot(122)
plt.bar(sorted_pop_region, sorted_pop ,color='lightgreen')
plt.xticks(rotation=65)
plt.title("Population by Region")
plt.ylabel("Population (billion)")

plt.show()
```
