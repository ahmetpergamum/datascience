### Q1
Bu soruda, merkezi limit teoreminin çalışıp çalışmadığını test edeceksiniz. İki adet normal dağılıma sahip 1000 gözlem içeren değişken üreteceksiniz. Bu değişkenlerin, ortalama ve standart sapmasını kendiniz belirleyebilirsiniz. Yapmanız gereken, ilk değişkeni 50 kere üretin ve her seferinde bu değişkenin ortalamasını alın. İkinci değişkeni ise 1000 kere üretin ve yine her seferinde bu değişkenin ortalamasını alın. Daha sonra, iki değişkenin hesapladığınız ortalamalarının histogramını çizdirin. Hangi değişkenin ortalamalarının dağılımı normal dağılıma daha yakın? Sizce Merkezi Limit Theoremi çalışmış görünüyor mu?

### Solution
1000 defa uretilen degiskenin ortalamasinin dagilimi normal dagilima daha yakindir.
Merkezi limit teoreminin etkisi gorulmektedir.

#### Merkezi Limit Teoremi:
Belirli bir popülasyondan tekrar tekrar alınan örneklem ortalamasının dağılımının şekli, örneklem büyüklükleri arttıkça, normal dağılıma yaklaşacaktır.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# custom font defs
title_font = {'family': 'times new roman', 'color': 'darkred','weight': 'bold','size': 22}
axis_font  = {'family': 'ubuntu mono', 'color': 'darkblue','weight': 'bold','size': 10 }

# create the normal distributed dataset 50 times and append the means to the list
meanOneList = []
for i in range(0,50):
        meanOneList.append(np.mean(np.random.default_rng().normal(10,2,1000)))

# create the normal distributed dataset 1000 times and append the means to the list
meanTwoList = []
for i in range(0,1000):
        meanTwoList.append(np.mean(np.random.default_rng().normal(50,2,1000)))


plt.figure(figsize=(24,8), dpi = 200)
plt.subplot(121)
plt.title('Data produced with 50 iterations', fontdict = title_font)
plt.hist(meanOneList,50)
plt.subplot(122)
plt.title('Data produced with 1000 iterations', fontdict = title_font)
plt.hist(meanTwoList,50)
plt.show()
```

### Q2
İki madeni paramızın olduğunu varsayalım. Bunlardan bir tanesi adil (yani, yeteri kadar deneme yapıldığında yazı veya tura 1/2 olasılığına yaklaşmakta), diğeri ise adil değil ve atıldığında 3/4 olasılığında tura elde edilmektedir. Herhangi biri seçilip, n kere deneme yapıldığında, Büyük Sayılar Yasası uzun vadede gelecek tura sayısını tahmin etmemize izin verir mi? Çok sayıda deneme yapıldığında, hangi paranın seçildiğini bilebilir miyiz?

### Solution

Buyuk sayilar yasasina gore uzun vadede gelecek tura sayisini tahmin edebiliriz.

Cok sayida deneme yapilirsa hangi paranin secildigini bilebiliriz.

```python
import numpy as np

# 1000 samples for each coin toss
fair = np.random.binomial(n=1, p=0.5, size=1000)
unfair = np.random.binomial(n=1, p=0.75, size=1000)

plt.figure(figsize=(24,8), dpi = 200)
plt.subplot(121)
plt.title('Fair Coin ', fontdict = title_font)
plt.hist(fair, color = 'lightgreen')
plt.subplot(122)
plt.title('Unfair Coin ', fontdict = title_font)
plt.hist(unfair, color = 'red')
plt.show()
```
